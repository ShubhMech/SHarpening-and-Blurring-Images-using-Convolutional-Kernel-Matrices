```python
gray_img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap= "gray")
```

### Sharpening Images

```python
sharpening=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
```

```python
sharpened_img= cv2.filter2D(gray_img,-1,sharpening)
```

```python
plt.imshow(sharpened_img, cmap="gray")
```

```python
sharpening1=np.array([[0,-1,0],[-1,9,-1],[0,-1,0]])
```

```python
sharpened_img1= cv2.filter2D(gray_img,-1,sharpening)
```

```python
plt.imshow(sharpened_img1, cmap="gray")
```

### Blurring Images

```python
blurring=np.array([[1,1,1],[1,1,1],[1,1,1]])
```

```python
blurring.shape
```

```python
blurring=(1/9)*blurring
```
